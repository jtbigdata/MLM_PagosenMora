## Información del Conjunto de Datos

Este conjunto de datos contiene información sobre pagos en mora, factores demográficos, datos de crédito, historial de pagos y declaraciones de facturación de clientes de tarjetas de crédito en Taiwán desde abril de 2005 hasta septiembre de 2005. [base de datos original](https://www.kaggle.com/datasets/uciml/default-of-credit-card-clients-dataset)

## Columnas de la base de datos

Hay 25 variables:

- **ID:** ID de cada cliente
- **LIMIT_BAL:** Monto del crédito otorgado en dólares NT (incluye crédito individual y familiar/suplente)
- **SEX:** Género (1=masculino, 2=femenino)
- **EDUCATION:** (1=escuela de posgrado, 2=universidad, 3=escuela secundaria, 4=otros, 5=desconocido, 6=desconocido)
- **MARRIAGE:** Estado civil (1=casado, 2=soltero, 3=otros)
- **AGE:** Edad en años
- **PAY_0:** Estado de pago en septiembre de 2005 (-1=pago a tiempo, 1=retraso en el pago por un mes, 2=retraso en el pago por dos meses, … 8=retraso en el pago por ocho meses, 9=retraso en el pago por nueve meses o más)
- **PAY_2:** Estado de pago en agosto de 2005 (escala igual a la anterior)
- **PAY_3:** Estado de pago en julio de 2005 (escala igual a la anterior)
- **PAY_4:** Estado de pago en junio de 2005 (escala igual a la anterior)
- **PAY_5:** Estado de pago en mayo de 2005 (escala igual a la anterior)
- **PAY_6:** Estado de pago en abril de 2005 (escala igual a la anterior)
- **BILL_AMT1:** Monto de la declaración de facturación en septiembre de 2005 (dólares NT)
- **BILL_AMT2:** Monto de la declaración de facturación en agosto de 2005 (dólares NT)
- **BILL_AMT3:** Monto de la declaración de facturación en julio de 2005 (dólares NT)
- **BILL_AMT4:** Monto de la declaración de facturación en junio de 2005 (dólares NT)
- **BILL_AMT5:** Monto de la declaración de facturación en mayo de 2005 (dólares NT)
- **BILL_AMT6:** Monto de la declaración de facturación en abril de 2005 (dólares NT)
- **PAY_AMT1:** Monto del pago previo en septiembre de 2005 (dólares NT)
- **PAY_AMT2:** Monto del pago previo en agosto de 2005 (dólares NT)
- **PAY_AMT3:** Monto del pago previo en julio de 2005 (dólares NT)
- **PAY_AMT4:** Monto del pago previo en junio de 2005 (dólares NT)
- **PAY_AMT5:** Monto del pago previo en mayo de 2005 (dólares NT)
- **PAY_AMT6:** Monto del pago previo en abril de 2005 (dólares NT)
- **default.payment.next.month:** Pago en mora (1=sí, 0=no)

**1. Carga y Preprocesamiento de Datos**

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
```

```python
# Separar las características y la variable objetivo
X = df.drop('default.payment.next.month', axis=1)
y = df['default.payment.next.month']

# Dividir el dataset en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Normalizar las características
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Balanceo de clases usando SMOTE
smote = SMOTE(random_state=42)
X_train, y_train = smote.fit_resample(X_train, y_train)

# Mostrar la nueva distribución de la variable objetivo en el conjunto de entrenamiento
print("Nueva distribución de la variable objetivo en el conjunto de entrenamiento:")
print(pd.Series(y_train).value_counts())
```

**2 Regresión Logística**

**2.1. Entrenamiento y Evaluación del Modelo**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Crear y entrenar el modelo de Regresión Logística
log_reg = LogisticRegression(random_state=42)
log_reg.fit(X_train, y_train)

# Hacer predicciones
y_pred_log_reg = log_reg.predict(X_test)
```

```python
# Evaluar el modelo de Regresión Logística
print(f'Accuracy de Regresión Logística: {accuracy_score(y_test, y_pred_log_reg)}')
print('Classification Report de Regresión Logística:')
print(classification_report(y_test, y_pred_log_reg))
```

```python
# Coeficientes de la Regresión Logística
coefficients = log_reg.coef_[0]
features = X.columns
coeff_df = pd.DataFrame({'Feature': features, 'Coefficient': coefficients})
coeff_df = coeff_df.sort_values(by='Coefficient', ascending=False)
```

```python
# Graficar los coeficientes
plt.figure(figsize=(8, 4))
sns.barplot(x='Coefficient', y='Feature', data=coeff_df)
plt.title('Coeficientes de Regresión Logística')
plt.xlabel('Coeficiente')
plt.ylabel('Característica')
plt.show()
```

**2.2. Optimización del Modelo**

```python
from sklearn.model_selection import GridSearchCV

# Definir los parámetros a ajustar
param_grid = {'C': [0.001, 0.01, 0.1, 1, 10, 100]}

# Crear y ajustar el modelo de Regresión Logística con búsqueda de hiperparámetros
grid_search_log_reg = GridSearchCV(LogisticRegression(random_state=42), param_grid, cv=5, scoring='accuracy')
grid_search_log_reg.fit(X_train, y_train)
```

```python
# Mostrar el mejor parámetro y el rendimiento del modelo
print(f'Best Parameters: {grid_search_log_reg.best_params_}')
print(f'Best Score: {grid_search_log_reg.best_score_}')
```

```python
# Evaluar el modelo optimizado
y_pred_log_reg_opt = grid_search_log_reg.best_estimator_.predict(X_test)
print(f'Optimized Accuracy de Regresión Logística: {accuracy_score(y_test, y_pred_log_reg_opt)}')
print('Optimized Classification Report de Regresión Logística:')
print(classification_report(y_test, y_pred_log_reg_opt))
```

**3. Random Forest**

**3.1. Entrenamiento y Evaluación del Modelo**

```python
from sklearn.ensemble import RandomForestClassifier

# Crear y entrenar el modelo Random Forest
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)

# Hacer predicciones
y_pred_rf = rf.predict(X_test)

# Evaluar el modelo Random Forest
print(f'Accuracy de Random Forest: {accuracy_score(y_test, y_pred_rf)}')
print('Classification Report de Random Forest:')
print(classification_report(y_test, y_pred_rf))
```

**3.2. Optimización del Modelo**

```python
from sklearn.model_selection import GridSearchCV

# Definir los parámetros a ajustar con menos combinaciones
param_grid_rf = {
    'n_estimators': [50,100],  # Menos valores para reducir el tiempo de búsqueda
    'max_depth': [None,5,10,15],  # Menos profundidades para reducir la complejidad
    'min_samples_split': [2, 5]  # Menos valores para reducir la complejidad
}
```

```python
# Crear y ajustar el modelo Random Forest con búsqueda de hiperparámetros
# Usar menos pliegues en la validación cruzada
grid_search_rf = GridSearchCV(RandomForestClassifier(random_state=42), param_grid_rf, cv=5, scoring='accuracy', n_jobs=-1)
```

```python
grid_search_rf.fit(X_train, y_train)
```

```python
# Mostrar el mejor parámetro y el rendimiento del modelo
print(f'Best Parameters: {grid_search_rf.best_params_}')
print(f'Best Score: {grid_search_rf.best_score_}')
```

```python
# Evaluar el modelo optimizado
y_pred_rf_opt = grid_search_rf.best_estimator_.predict(X_test)
print(f'Optimized Accuracy de Random Forest: {accuracy_score(y_test, y_pred_rf_opt)}')
print('Optimized Classification Report de Random Forest:')
print(classification_report(y_test, y_pred_rf_opt))
```

**Comparación de Modelos: Regresión Logística vs Random Forest**

**Métricas de Evaluación**

**Accuracy (Exactitud)**
- **Definición**: Proporción de predicciones correctas sobre el total de predicciones realizadas.
- **Fórmula**: $\text{Accuracy} = \frac{\text{Número de predicciones correctas}}{\text{Número total de predicciones}}$
- **Interpretación**: Mide la capacidad general del modelo para hacer predicciones correctas.

**Classification Report (Informe de Clasificación)**
- **Precision (Precisión)**: Proporción de verdaderos positivos sobre el total de positivos predichos.
  - **Fórmula**: $\text{Precision} = \frac{TP}{TP + FP}$
  - **Interpretación**: Indica la calidad de las predicciones positivas.
- **Recall (Exhaustividad)**: Proporción de verdaderos positivos sobre el total de positivos reales.
  - **Fórmula**: $ \text{Recall} = \frac{TP}{TP + FN} $
  - **Interpretación**: Mide la capacidad del modelo para encontrar todos los ejemplos positivos.
- **F1-Score**: Media armónica entre precisión y recall.
  - **Fórmula**: $ F1 = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}} $
  - **Interpretación**: Proporciona una métrica balanceada entre precisión y recall.

**Macro Average**
- **Definición**: Promedia las métricas de cada clase sin tener en cuenta el soporte de cada clase.
- **Interpretación**: Da una idea general del rendimiento del modelo en todas las clases.

**Weighted Average**
- **Definición**: Promedia las métricas de cada clase ponderadas por el número de verdaderos ejemplos en cada clase.
- **Interpretación**: Refleja el desempeño del modelo teniendo en cuenta el desequilibrio de clases.

**Recall y F1-Score**
- **Regresión Logística**:
  - **Recall en Clase 0**: El recall para la clase 0 es de 0.69, indicando que el modelo detecta el 69% de los verdaderos negativos. Esto es bueno, pero muestra que hay un 31% de casos negativos que el modelo no está identificando.
  - **Recall en Clase 1**: El recall para la clase 1 es de 0.66 en el modelo original y 0.65 en el optimizado. Sin embargo, el F1-score para la clase 1 es bajo (0.47), lo que refleja una combinación no ideal de precisión y recall para esta clase. Esto se debe a que la precisión en la clase 1 también es baja.

- **Random Forest**:
  - **Recall en Clase 0**: Con un recall de 0.89, el Random Forest tiene un excelente desempeño en identificar los verdaderos negativos. Esto sugiere que el modelo es más confiable para detectar casos donde no hay default.
  - **Recall en Clase 1**: El recall para la clase 1 es de 0.46, que es menor en comparación con la Regresión Logística. Aunque el Random Forest logra una mejor precisión para la clase 1, el recall sigue siendo limitado. El F1-score para la clase 1 es de 0.49, que también es una mejora sobre el modelo de Regresión Logística.

**Balance de Variables**
- **Regresión Logística**: La Regresión Logística enfrenta dificultades con el desbalance de clases, como lo muestran las bajas métricas para la clase 1. Aunque optimizar el modelo ayuda, sigue siendo un desafío manejar las desigualdades en los datos.

- **Random Forest**: El Random Forest, siendo un modelo basado en ensamble, es más robusto frente al desbalance de clases. Su capacidad para manejar interacciones complejas entre variables y su robustez frente al overfitting contribuyen a un mejor rendimiento en general.

**Conclusión**
El Random Forest demuestra un desempeño superior en términos de precisión general (Accuracy) y recuperación (Recall) para la clase 0, así como un mejor F1-score para la clase 1 en comparación con la Regresión Logística. A pesar de que el Random Forest no resuelve completamente el problema de desbalance de clases, su capacidad para manejar datos complejos y su mejor rendimiento en general lo hacen una opción más adecuada para este conjunto de datos. La Regresión Logística, aunque útil, podría no ser suficiente en problemas donde la identificación precisa de casos positivos es crucial, y podría beneficiarse de técnicas adicionales para mejorar su desempeño en la clase positiva.
